```python
def transform_to_integer(i) :
    le = preprocessing.LabelEncoder()
    le.fit(data.iloc[:,i])
    data.iloc[:,i] = le.transform(data.iloc[:,i])
```

```python
not_numbers = [1,3,4,5,6,9,10,11,12,14]
```

```python
for i in not_numbers:
    transform_to_integer(i)
```
