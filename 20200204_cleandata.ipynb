```python
def Transmission(x):
    if type(x)==str:
        if 'automatic' in x.lower():
            return 'Automatic'
        elif 'manual' in x.lower():
            return 'Manual'
        elif 'CVT' in x:
            return 'CVT'
        elif 'Dual Clutch' in x:
            return 'Dual Clutch'
        elif 'Continuously' in x:
            return 'CVT'
        else:
            return x
    else:
        return x
# 20200116
```

```python
data['Transmission'] = data['Transmission'].apply(Transmission)
```

```python
def trans_engine(x):
    if type(x)==str:
        if 'V8' in x:
            return 'V8'
        elif 'V6' in x:
            return 'V6'
        elif 'I4' in x:
            return 'I4'
        elif 'I6' in x:
            return 'I6'
        elif 'I5' in x:
            return 'I5'
        elif 'I3' in x:
            return 'I3'
        elif 'V12' in x:
            return 'V12'
        elif 'W12' in x:
            return 'W12'
        elif 'H4' in x:
            return 'H4'
        elif 'V10' in x:
            return 'V10'
        else:
            return x
# 20200116
```

```python
data['Engine'] = data['Engine'].apply(trans_engine)
```
