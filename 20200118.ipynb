```python
columns_need_std =['Price',
                   'Mileage',
                   'SellingDays','OriginPrice','Median Household Income(USD)','Density','Pop','LandArea','Vehicle Count']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std = StandardScaler()
```

```python
std.fit_transform(data[columns_need_std])
```

```python
data[columns_need_std] = std.fit_transform(data[columns_need_std])
```

```python
data.head(10)
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=0.2, min_samples=10, metric='euclidean')
```

```python
y_db = db.fit_predict(data)
```

```python
y_db
```

```python
np.unique(y_db)
```

```python
labels = db.labels_
```

```python
np.unique(labels)
```

```python
len(db.components_[0])
```
